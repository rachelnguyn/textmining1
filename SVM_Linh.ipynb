```python
import json

itemList = []
print("Started Reading JSON file: ")
with open('Luxury_Beauty_5.json') as f:
    for jsonObj in f:
        itemDict = json.loads(jsonObj)
        itemList.append(itemDict)

df = pd.DataFrame(itemList) 
df2 = df.drop(["verified", "style","unixReviewTime", "reviewerName","vote", "image"], axis=1)
df2 = df2.fillna('')
df2
```

```python
test_reviews = np.array(df2['reviewText'])
train_reviews = np.array(df2['reviewText'])
```

```python
senti= df2[df2["overall"].notnull()]

senti["senti"] = senti["overall"]>=4
senti["senti"] = senti["senti"].replace([True , False] , ["pos" , "neg"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range = (1,2)) 
feature_matrix_TRAIN = vectorizer.fit_transform(train_reviews).astype(float)
```

```python
feature_matrix_TRAIN_names = vectorizer.get_feature_names() 
feature_matrix_TRAIN_table = pd.DataFrame(data = feature_matrix_TRAIN.todense(), 
                                          columns = feature_matrix_TRAIN_names)
feature_matrix_TRAIN_table.head()
```

```python
feature_matrix_TEST = vectorizer.transform(test_reviews)
```

```python
df2["senti"] = senti["senti"].replace([True , False] , ["positive" , "negative"])
```

```python
test_polarity = np.array(df2['senti'])
train_polarity = np.array(df2['senti'])
```

```python
from sklearn import linear_model

svm = linear_model.SGDClassifier(loss='hinge', random_state = 0) 
svm.fit(feature_matrix_TRAIN, train_polarity)
predicted_svm = svm.predict(feature_matrix_TEST) 
```

```python
from sklearn import metrics
print('Accuracy rate:', np.round(metrics.accuracy_score(test_polarity, predicted_svm), 3))
```
